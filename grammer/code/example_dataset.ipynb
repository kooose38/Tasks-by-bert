```python
import pickle 
from pprint import pprint 
import pandas as pd

dataset = pd.DataFrame([])
c, w = [], []
with open("./dataset/test.txt", "rb") as f:
    test = pickle.load(f)
    test = test[:10]
    for d in test:
        correct = d["correct_text"]
        wrong = d["wrong_text"]
        c.append(correct)
        w.append(wrong)
dataset["correct_text"] = c
dataset["wrong_text"] = w 
        
dataset
```
