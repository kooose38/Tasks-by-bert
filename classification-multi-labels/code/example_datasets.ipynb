### Sample dataset 

```python
import pickle 
import pandas as pd 
f = open("./dataset/dataset.txt", "rb")
data = pickle.load(f)

sentence, labels = [], []
for d in data:
    text = d["text"]
    label = d["labels"]
    sentence.append(text)
    labels.append(label)
    
df = pd.DataFrame(labels, columns=["negative", "neutral", "positive"])
df["text"] = sentence 
df[["text", "negative", "neutral", "positive"]].sample(10)
```
